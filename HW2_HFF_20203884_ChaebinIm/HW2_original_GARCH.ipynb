```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" # 한 실행칸에 프린트 여러개 해도 다 출력시키도록 하는 코드.
```

```python
data = pd.read_csv('./data_hw2_preprocessed.csv')
data2 = pd.read_csv('./GARCH_Itos.csv')
data.head()
```

```python
data2.head()
```

```python
from arch import arch_model
```

```python
GARCH_res = []
for i in range(80, len(data)):
    am = arch_model(data['daily_ret'][:i]*data['daily_ret'][:i], vol='Garch', p=1, o=0, q=1, dist='Normal')
    res = am.fit(update_freq=1)
    GARCH_res.append(res.forecast().mean.iloc[-1][0])
```

```python
data2.head()
```

```python
data2['GARCH'] = GARCH_res
data2['real'] = data['realized_vol'].iloc[80:].reset_index()['realized_vol']
data2.head()
```

```python
plt.figure(figsize = (50, 20))
data2.plot()
plt.show()
```

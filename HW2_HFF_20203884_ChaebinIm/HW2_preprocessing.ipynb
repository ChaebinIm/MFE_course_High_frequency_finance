```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" # 한 실행칸에 프린트 여러개 해도 다 출력시키도록 하는 코드.
```

```python
data = pd.read_csv('kospi_1분봉_201910_202010.csv')
data.head()
data.tail()
```

```python
data.columns = ['date', 'time', 'open', 'high', 'low', 'close', 'volume', 'a', 'b']
data['datetime'] = data['date'] + ' ' + data['time']
```

```python
date_list = list(data['date'].unique())[1:]
```

```python
date = data['date'].unique()
```

```python
data = data[['date', 'datetime', 'close']]
data.head()
```

```python
data['datetime'] = pd.to_datetime(data['datetime'], format = '%Y%m%d %H:%M:%S')
data.head()
data.dtypes
```

```python
daily_ret = []
for i in date_list:
    tmp = data[data['date'] == i]['close'].reset_index()['close']
    daily_ret.append((i, np.log(tmp.iloc[-1]/tmp.iloc[0])))
daily_ret
```

```python
begin_date = data['datetime'][0].date()
end_date = data['datetime'][len(data)-1].date()
begin_date
end_date
```

```python
data['close'] = np.log(data['close'] / data['close'].shift(1))
```

```python
data = data[data['datetime'] >= '2019-10-31'].reset_index(drop = True)
data.columns = ['date', 'datetime', 'ret']
data.head()
```

```python
daily_ret
```

```python
real_vol = []
for _date in date_list:
    tmp_ret = data[data['date'] == _date]['ret']
    tmp_ret = tmp_ret * tmp_ret
    res = tmp_ret.sum()
    real_vol.append((_date, res))
real_vol
```

```python
daily_ret = np.array(daily_ret)
real_vol = np.array(real_vol)
len(daily_ret)
len(real_vol)
```

```python
final_data = pd.DataFrame(data = np.array([daily_ret[:,0], daily_ret[:,1], real_vol[:, 1]]).T, columns = ['date', 'daily_ret', 'realized_vol']).iloc[:-1,:]
```

```python
final_data.to_csv('data_hw2_preprocessed.csv')
```
